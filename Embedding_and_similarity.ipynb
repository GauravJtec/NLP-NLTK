Cosine Similarity

```python
# Compute cosine similarity between all pairs

sentences = ['A woman is reading a book in the park.',
        'A boy is playing soccer with his friends in the backyard.',
        'A group of people are enjoying a picnic near the lake.',
        'A man is walking his dog in the park.',
        'A woman is watering the plants in her garden.',
        'A child is playing with a toy car on the floor.',
        'A man is cleaning his car in the driveway.',
        'A group of friends are hiking in the mountains.',
        'A woman is shopping for groceries at the supermarket.',
        'A man is jogging on the beach.',
        'A group of tourists are taking pictures in front of a famous landmark.'
          ]

embeddings=model.encode(sentences)
#Compute cosine similarity between all pairs
cos_sim= util.cos_sim(embeddings,embeddings)
```

```python
cos_sim
```

```python
combination=[]
for i in range(len(cos_sim)-1):
  for j in range(i+1,len(cos_sim)):
      combination.append((cos_sim[i][j],i,j))
combination
```

```python
#Sort list by the highest cosine similarity score

combination=sorted(combination, key=lambda x:x[0],reverse=True)
```

```python
print("Top-5 most similar pairs:")

for score,i,j in combination[0:5]:
  print('{} \t {} \t {:.4f}'.format(sentences[i],sentences[j],cos_sim[i][j]))
```

Semantic Search

```python
from sentence_transformers import SentenceTransformer,util
model=SentenceTransformer('clips/mfaq')
```

```python
Question= 'How do I reset my password?'

Answer_1= 'To reset your password, go to the login page and click on the "forgot password" link. Follow the instructions to reset your password.'

Answer_2= 'If you\'ve forgotten your password, you can reset it by clicking on the "forgot password" link on the login page.You\'ll be prompted to enter your email address and follow the instructions to reset your password.'

Answer_3 ='If you need to reset your password, go to the login page and click on the "forgot password" link. Enter your email address and follow the instructions to reset your password.'


ques_embedding=model.encode(Question)
ans_embeddings=model.encode([Answer_1,Answer_2,Answer_3])

print(util.semantic_search(ques_embedding,ans_embeddings))
```

Clustering

```python
from sklearn.cluster import KMeans
import numpy as np

embedder = SentenceTransformer('all-MiniLM-L6-v2')

sentences = ['A woman is reading a book in the park.',
        'A boy is playing soccer with his friends in the backyard.',
        'A group of people are enjoying a picnic near the lake.',
        'A man is walking his dog in the park.',
        'A woman is watering the plants in her garden.',
        'A child is playing with a toy car on the floor.',
        'A man is cleaning his car in the driveway.',
        'A group of friends are hiking in the mountains.',
        'A woman is shopping for groceries at the supermarket.',
        'A man is jogging on the beach.',
        'A group of tourists are taking pictures in front of a famous landmark.'
          ]
totalembeddings=embedder.encode(sentences)
```

```python
# Normalize the embeddings to unit length
totalembeddings = totalembeddings /  np.linalg.norm(totalembeddings, axis=1, keepdims=True)
     
```

```python
totalembeddings[0]
```

```python
clustering_model=KMeans(n_clusters=4)
clustering_model.fit(totalembeddings)
cluster_assignment = clustering_model.labels_
print(cluster_assignment)
```

```python

clustered_sentences = {}
for sentence_id, cluster_id in enumerate(cluster_assignment):
    if cluster_id not in clustered_sentences:
        clustered_sentences[cluster_id] = []

    clustered_sentences[cluster_id].append(sentences[sentence_id])
clustered_sentences
     
```
